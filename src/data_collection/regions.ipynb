```python
import camelot

# Read the PDF file
tables = camelot.read_pdf('C:\\Users\\abdel\\Desktop\\Implantation Bancaire 2023 VF.pdf', pages='9-end')
```

```python
regions = ["Tangier-Tetouan-Al Hoceima",
           "L'Oriental",
           "Fez-Meknes",
           "Rabat-Salé-Kénitra",
           "Béni Mellal-Khénifra",
           "Casablanca-Settat",
           "Marrakech-Safi",
           "Drâa-Tafilalet",
           "Souss-Massa",
           "Guelmim-Oued Noun",
           "Laâyoune-Sakia El Hamra",
           "Dakhla-Oued Ed-Dahab"]
```

```python
import pandas as pd
tables[0].df
```

```python
# Step 3: Map cities to regions
result_dict = {}

for table,region in zip(tables,regions):
    df = table.df
    # Convert column 28 (index 28) to numeric, handling errors by coercing invalid values to NaN
    df.iloc[:, 28] = pd.to_numeric(df.iloc[:, 28], errors='coerce')
    # Step 2: Filter cities with more than 10 banks
    filtered_cities = df[df.iloc[:, 28] > 10]  # Column 28 is index 28 (0-based)
    filtered_cities = filtered_cities.iloc[:, 0].tolist()  # Column 1 is index 0 (0-based)
    result_dict[region]=filtered_cities
```

```python
result_dict
```

```python
import json
# Step 1: Write the dictionary to a JSON file
output_file = './cities.json'  # Name of the output JSON file

with open(output_file, 'w') as json_file:
    json.dump(result_dict, json_file, indent=4)  # indent=4 for pretty formatting

print(f"Dictionary has been written to {output_file}")
```
